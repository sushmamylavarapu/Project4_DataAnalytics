```python
import requests
import json
import pandas as pd
import re
from pprint import pprint

# Import the API key
from config import omdb_key, tmdb_token
```

```python
# read oscar winning movie list
movie_df = pd.read_csv("./Resources/TMDB10000.csv", index_col='Unnamed: 0')

# remove blank film names
movie_df = movie_df.dropna()

movie_df
```

```python
tmdb_data = []

for i in range(len(movie_df)):
    
    if i % 2 == 1 : # select 5000 movies to export data
        
        try: 

            # get data by tmdb_id
            temp_id = movie_df.loc[ i , 'id' ]

            ## read json from tmdb api
            tmdb_url = f"https://api.themoviedb.org/3/movie/{temp_id}?language=en-US"
            headers = {
                "accept": "application/json",
                "Authorization": tmdb_token}

            tmdb_response = requests.get( tmdb_url, headers=headers ).json()

            omdb_id = tmdb_response['imdb_id']

            ## read json from omdb api
            omdb_url = f"http://www.omdbapi.com/?apikey={omdb_key}&i={omdb_id}" # by imdb_id ie. tt0112870

            omdb_response = requests.get( omdb_url ).json()

            tmdb_response[ 'imdb_data' ] = omdb_response

            tmdb_data.append(tmdb_response)
        
        except :
            print(tmdb_response)
            print(omdb_response)
            pass

    if i % 500 == 0 :
        print(f'{i} / { len(movie_df) }')
    
print(f'-----complete {len(tmdb_data)} movies-----')
```

```python
# save data
with open("./Resources/movie_data.json", "w") as outfile:
    json.dump(tmdb_data, outfile)
```
